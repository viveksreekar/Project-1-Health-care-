```python
import gradio as gr
import numpy as np
from PIL import Image
from tensorflow import keras

# Load your trained model
model = keras.models.load_model("model.keras")

# Class labels
classes = ['Real', 'Fake']

# Image preprocessing
def preprocess_image(image):
    image = image.resize((224, 224))
    image = np.array(image).astype("float32") / 255.0
    image = np.expand_dims(image, axis=0)
    return image

# Prediction function
def predict(image):
    if image is None:
        return "Error: No image received."
    
    try:
        img_array = preprocess_image(image)
        prob_fake = float(model.predict(img_array)[0][0])
        prob_real = 1 - prob_fake

        return (
            f"Prediction:\n"
            f"Real: {prob_fake * 100:.2f}%\n"
            f"Fake: {prob_real * 100:.2f}%"
        )
    except Exception as e:
        return f"Error during prediction: {str(e)}"

# Gradio interface
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="pil", label="Upload an Image"),
    outputs="text",
    title="Fake vs Real Image Detector",
    description="Uploads an image and returns confidence scores for both Fake and Real."
)

iface.launch()
```
